```python
df = pd.read_csv("../00_datasets/titanic.csv")
df.sample(5)
```

```python
# bar plot using plt
df['sex'].value_counts().plot(kind='bar')
```

```python
# creata a bar plot using matplotlib
plt.bar(df['who'], df['age'], color='maroon')
```

```python
# Figure Size
fig = plt.figure(figsize =(7, 4)) # width, height
# creata a bar plot using matplotlib
plt.bar(df['who'], df['age'], color='maroon')
# save the plot in png with 300 dpi resolution
plt.savefig('./01_plots/bar_plot.png', dpi=300)
# save svg file
plt.savefig('./01_plots/bar_plot_svg.svg', format='svg')

# show the plot
plt.show()
```

```python
fig = plt.figure(figsize =(7, 4)) # width, height
plt.scatter(df['age'], df['fare'], color='maroon')
# label plots
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
# creata a bar plot using matplotlib
plt.stem(df['who'], df['age'])
```

```python
plt.stackplot(df['who'], df['age'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.stairs(y, linewidth=2.5)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig = plt.figure(figsize =(7, 4)) # width, height
sns.boxplot(data = df, x = 'sex', y='age')
# add labels
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Box Plot")
plt.savefig('./01_plots/box_plot_seaborn.png', dpi=300)
```

```python
sns.boxenplot(data = df, x = 'sex', y='age')
```

```python
fig = plt.figure(figsize =(7, 4)) # width, height
# change the colors based on the palette
sns.set_palette("Set1")
sns.barplot(data = df, x = 'pclass', y='age', hue='sex')
```

```python
fig = plt.figure(figsize =(7, 4)) # width, height
# change the colors based on the palette
sns.set_palette("Set1")
sns.boxplot(data = df, x = 'pclass', y='age', hue='sex')
```

```python
fig = plt.figure(figsize =(7, 4)) # width, height
# change the colors based on the palette
sns.set_palette("Set1")
sns.boxenplot(data = df, x = 'class', y='age', hue='sex', 
              order=['First', 'Second', 'Third'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Load the example diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
```
