```python
import seaborn as sns

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic_data.head()
```

```python
sns.countplot(x='class', data=titanic_data)
```

```python
# Count unique values for each column
unique_values = titanic_data.nunique()

# Plot the unique values
unique_values.plot(kind='bar', figsize=(12, 6), title='Unique Values in Titanic Dataset')
```

```python
import plotly.express as px

# Plot 1: Bar plot of the number of passengers in each class
fig1 = px.bar(titanic_data, x='class', title='Number of Passengers in Each Class')
fig1.show()

# Plot 2: Histogram of the age distribution
fig2 = px.histogram(titanic_data, x='age', nbins=30, title='Age Distribution of Passengers')
fig2.show()

# Plot 3: Box plot of fare prices by class
fig3 = px.box(titanic_data, x='class', y='fare', title='Fare Prices by Class')
fig3.show()

# Plot 4: Scatter plot of age vs fare, colored by survival
fig4 = px.scatter(titanic_data, x='age', y='fare', color='survived', title='Age vs Fare (Colored by Survival)')
fig4.show()
```

```python
import plotly.express as px

# Create a polar coordinate plot for the 'fare' and 'age' columns
fig_polar = px.scatter_polar(titanic_data, r='fare', theta='age', color='survived', title='Polar Coordinate Plot of Fare vs Age (Colored by Survival)')
fig_polar.show()
```

```python
# creata a dataset of titanic_data calculating the average age and fare for each class
average_age_fare = titanic_data.groupby('class')[['age', 'fare']].mean().reset_index()
# Plot the average age and fare for each class
fig_avg = px.bar(average_age_fare, x='class', y=['age', 'fare'], title='Average Age and Fare for Each Class', barmode='group')
fig_avg.show()
```
