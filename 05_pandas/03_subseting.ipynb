# Subsetting the data

1. Selection based on the column names
2. Selection based on the rows for a specific column selection criteria
3. Selection based on the rows for multiple column selection criteria

```python
# import library
import pandas as pd

df = pd.read_csv('../00_datasets/titanic.csv')
df
```

```python
# subset age column fromt the data

age = df['age']
age
type(age)
```

```python
gender = df['sex']
gender
```

```python
# subset two column named age and sex
df_age_sex = df[["age", "sex"]]
print(type(df_age_sex))
df_age_sex
```

```python
# filter
age_filter = df['age'] > 18
age_filter
```

```python
# filter the rows based on age column
adults = df[df['age'] >= 18]
non_adults = df[df['age'] < 18]

adults.to_csv('../00_datasets/pandas_practice/adults_titanic_filter.csv', index=False)
```

```python
# filter the data of adult males from the df sex==male and age>=18
adult_males = df[(df['sex'] == "male") & (df['age'] >= 18)]
adult_males
# adult_males[['sex', 'age']]
```

```python
# filter the data and retain only two columns
df_01 = df[['sex', 'age']]
df_01 = df_01[(df_01['sex'] == "male") & (df_01['age'] >= 18)]
df_01
```

```python
# filter the data of adult males from the df sex==female and age>=18
adult_females = df[(df['sex'] == "female") & (df['age'] >= 18)]
adult_females
```

# Subset if something is known/unknown

```python
# import library
import pandas as pd
df = pd.read_csv('../00_datasets/titanic.csv')
df
```

```python
df_age_known = df[df['age'].notna()] # or use notnull()
df_age_known
```

```python
df_age_unknown = df[df['age'].isna()] # or use isnull()
df_age_unknown
```
